### Tokenization

There are multiple different ways to build a tokenizer:

#### Tokenization Training Methods

**Byte-Pair Encoding (BPE):**
Iteratively merges the most frequent character pairs in a corpus to create subword tokens. Efficient and widely used in models like GPT-2 and RoBERTa.

**WordPiece:**
Similar to BPE but chooses merges based on maximizing the likelihood of the data. Used in BERT and DistilBERT for effective handling of rare words.

**Unigram Language Model:**
Starts with a large set of subwords and prunes the vocabulary based on probabilistic modeling. Allows multiple tokenizations and is used in T5 and XLNet (via SentencePiece).

**Byte-Level BPE:**
Applies BPE to byte sequences, ensuring all text—including any language or character—can be tokenized. Robust and used in GPT-2, RoBERTa.

### Generating a Sentence Embedding Step-by-Step

```python
import torch.nn.functional as F
from torch import Tensor
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
```

#### 1. Tokenize Input

```python
# Tokenize the input texts
tokenized_inputs = tokenizer([sentence], max_length=512, padding=True, truncation=True, return_tensors='pt')
tokenized_inputs
```

```python
tokenizer.convert_ids_to_tokens(tokenized_inputs['input_ids'][0])
```

#### 2. Encode Tokenized Inputs

```python
contextualized_token_embeddings = model(**tokenized_inputs).last_hidden_state
contextualized_token_embeddings
```

```python
contextualized_token_embeddings.shape
```

#### 3. Average Contextualized Token Embeddings

```python
def average_pool(token_embeddings: Tensor, attention_mask: Tensor) -> Tensor:
    masked_token_embeddings = token_embeddings.masked_fill(~attention_mask[..., None].bool(), 0.0)
    return masked_token_embeddings.sum(dim=1) / attention_mask.sum(dim=1)[..., None]

sentence_embedding = average_pool(contextualized_token_embeddings, tokenized_inputs['attention_mask'])
sentence_embedding.shape
```

#### 4. Normalize Sentence Embedding

```python
# normalize embeddings
sentence_embedding = F.normalize(sentence_embedding, p=2, dim=1)
print(sentence_embedding.shape)
```
