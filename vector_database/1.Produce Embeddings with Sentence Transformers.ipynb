## How to Produce Embeddings with Sentence Transformers

how to generate sentence embeddings using pre-trained models from the Sentence Transformers library.

```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [
    # Group 1: Animals
    "The quick brown fox jumps over the lazy dog.",
    "A fast, dark-colored fox leaps above a sleepy canine.",
    "A sly fox swiftly moves past a tired hound.",
    "Dogs are loyal companions and love to play fetch.",
    "Foxes and dogs are both members of the canine family.",

    # Group 2: Technology/AI
    "Artificial intelligence is transforming the world.",
    "Machine learning algorithms are advancing rapidly.",
    "Deep learning is a powerful branch of artificial intelligence.",
    "Robots powered by AI are becoming more common in industry.",
    "Natural language processing helps computers understand text.",

    # Group 3: Food
    "Pizza is my favorite food to eat on weekends.",
    "I love making homemade pizza with fresh ingredients.",
    "Burgers and fries are a classic fast food combination.",
    "Sushi is a popular Japanese dish made with rice and fish.",
    "Many people enjoy ice cream as a dessert in the summer."
]
```

```python
# Generate embeddings for the sentences
embeddings = model.encode(sentences)

# Display the shape of the embeddings
print(f"Generated embeddings shape: {embeddings.shape}")
```

```python
embeddings[0]
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions for visualization
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)

# Plot the embeddings
plt.figure(figsize=(20,8))
for i, label in enumerate(sentences):
    x, y = reduced_embeddings[i]
    plt.scatter(x, y)
    plt.annotate(label, (x, y))
plt.title("PCA Projection of Sentence Embeddings")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

## Similarity Metrics — Cosine, Euclidean, Dot Product

Understand and compute different similarity metrics between sentence embeddings.

**Cosine Similarity:**

$$
\text{cosine_similarity}(\mathbf{a}, \mathbf{b}) = \frac{\mathbf{a} \cdot \mathbf{b}}{||\mathbf{a}|| \, ||\mathbf{b}||}
$$

**Euclidean Distance:**

$$
\text{euclidean_distance}(\mathbf{a}, \mathbf{b}) = ||\mathbf{a} - \mathbf{b}|| = \sqrt{\sum_{i=1}^{n} (a_i - b_i)^2}
$$

**Dot Product:**

$$
\text{dot_product}(\mathbf{a}, \mathbf{b}) = \sum_{i=1}^{n} a_i b_i
$$

```python
import numpy as np
from numpy.linalg import norm

def cosine_similarity(a, b):
    return np.dot(a, b) / (norm(a) * norm(b))

def euclidean_distance(a, b):
    return norm(a - b)

def dot_product(a, b):
    return np.dot(a, b)
```

```python
# Select pairs of sentences to compare
pairs = [
    (0, 1),  # Similar sentences
    (0, -1)   # Dissimilar sentences
]

# Compute and display similarity metrics
for idx1, idx2 in pairs:
    emb1 = embeddings[idx1]
    emb2 = embeddings[idx2]
    print(f"\nComparing:\n{sentences[idx1]}\n{sentences[idx2]}\n\n")
    print(f"Cosine Similarity: {cosine_similarity(emb1, emb2):.4f}")
    print(f"Euclidean Distance: {euclidean_distance(emb1, emb2):.4f}")
    print(f"Dot Product: {dot_product(emb1, emb2):.4f}")
    print("-" * 50)
```


## Using Embeddings to Create a Mini Search Engine

Build a simple semantic search engine using sentence embeddings.


```python
def semantic_search(query, corpus, corpus_embeddings, top_k=5):
    # Generate embedding for the query
    query_embedding = model.encode([query])[0]

    # Compute similarities
    similarities = [dot_product(query_embedding, doc_embedding) for doc_embedding in corpus_embeddings]

    # Get top_k results
    top_k_indices = np.argsort(similarities)[-top_k:][::-1]

    print(f"\nQuery: {query}\nTop {top_k} results:")
    for idx in top_k_indices:
        print(f"Score: {similarities[idx]:.4f} | Document: {corpus[idx]}")
```

```python
# Example query
query = "pets to adopt near me"

# Perform semantic search
semantic_search(query, sentences, embeddings, top_k=5)
```
